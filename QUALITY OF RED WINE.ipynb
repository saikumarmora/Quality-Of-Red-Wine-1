```python
import numpy as np
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

**linear regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import  r2_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Coefficients:', lr.coef_)
print('Intercept:', lr.intercept_)
print('Mean SquaredÂ Error:',mse)
r2 = r2_score(y_test, y_pred)
print('R-Squared:%.2f' % r2)
```

```python
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=3)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

**svm**

```python
from sklearn.svm import SVR
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
sv = SVR(kernel='linear')
sv.fit(X_train, y_train)
y_pred = sv.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Coefficients:', sv.coef_)
print('Intercept:', sv.intercept_)
print('Mean SquaredÂ Error:',mse)
r2 = r2_score(y_test, y_pred)
print('R-Squared:%.2f' % r2)
```

**knn**

```python
from sklearn.neighbors import KNeighborsRegressor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
pr = KNeighborsRegressor(n_neighbors=5)
pr.fit(X_train, y_train)
y_pred = pr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Mean SquaredÂ Error:',mse)
r2 = r2_score(y_test, y_pred)
print('R-Squared:%.2f' % r2)
```

 decision **tree**

```python
from sklearn.tree import DecisionTreeRegressor
X_train, X_test, y_train, y_test = train_test_split (X,y ,test_size=0.20, random_state=42)
dt = DecisionTreeRegressor(max_depth=7)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Mean SquaredÂ Error:',mse)
r2 = r2_score(y_test, y_pred)
print('R-Squared:%.2f' % r2)
```

***Random forest***

```python
from sklearn.ensemble import RandomForestRegressor
X_train, X_test, y_train, y_test = train_test_split(X,y ,test_size=0.20, random_state=42)
rf = RandomForestRegressor(n_estimators=100, max_depth=7)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Mean SquaredÂ Error:',mse)
r2 = r2_score(y_test, y_pred)
print('R-Squared:%.2f' % r2)
```

**ADA Boost**

```python
from sklearn.ensemble import AdaBoostRegressor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
dtr = DecisionTreeRegressor(max_depth=1)
abr = AdaBoostRegressor(n_estimators=100, learning_rate=0.01, base_estimator=dtr)
abr.fit(X_train, y_train)
y_pred = abr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Mean SquaredÂ Error:',mse)
r2 = r2_score(y_test, y_pred)
print('R-Squared:%.2f' % r2)
```

**Gradient Boosting**

```python
from sklearn.ensemble import GradientBoostingRegressor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
gbr = GradientBoostingRegressor(n_estimators=500, learning_rate=0.01, max_depth=3)
gbr.fit(X_train, y_train)
y_pred = gbr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:',mse)
r2 = r2_score(y_test, y_pred)
print('R-Squared:%.2f' % r2)
```

**XG Boost**

```python
import xgboost as xgb
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
xgbr = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3)
xgbr.fit(X_train, y_train)
y_pred = xgbr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:',mse)
r2 = r2_score(y_test, y_pred)
print('R-Squared:%.2f' % r2)
```
